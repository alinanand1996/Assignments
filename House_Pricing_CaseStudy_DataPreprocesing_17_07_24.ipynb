```python
df1.columns
```

```python

for col in df1.columns:
  q1=np.percentile(df[col],25,interpolation='midpoint')
  q2=np.percentile(df[col],50,interpolation='midpoint')
  q3=np.percentile(df[col],75,interpolation='midpoint')
  iqr=q3-q1
  upper=q3+1.5*iqr
  lower=q1-1.5*iqr
  outlies=[]
  for x in df[col]:
    if((x>upper) or (x<lower)):
      outlies.append(x)
  df[col]=df[col].clip(lower,upper)
```

```python
df['Date House was Sold']=pd.to_datetime(df['Date House was Sold'],format='%d %B %Y')
df['Date House was Sold']
df['HS_day']=df['Date House was Sold'].dt.day
df['HS_month']=df['Date House was Sold'].dt.month
df['HS_year']=df['Date House was Sold'].dt.year
df.drop(['Date House was Sold'],axis=1,inplace=True)
df.head()
```

```python
df.info()
```

```python
cat_cols=df.select_dtypes(include=['object'])
cat_columns=cat_cols.columns
cat_columns
```

```python
num_cols=df.select_dtypes(include=['int64','float64','int32'])
num_columns=num_cols.columns
num_columns
```

```python
data_encode=pd.get_dummies(df,columns=cat_columns,dtype=int)
data_encode.head()
```

```python
data_std_scale=data_encode.copy()
std_scale=StandardScaler()
data_std_scale[num_columns]=std_scale.fit_transform(data_std_scale[num_columns])
data_std_scale.head()
```

```python
corr_cols=['Sale Price','No of Bedrooms','Flat Area (in Sqft)','No of Floors','Overall Grade','Waterfront View_No','Waterfront View_Yes']
corr=data_std_scale[corr_cols].corr()
corr
```
